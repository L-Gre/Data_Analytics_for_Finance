### Group By

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "height (in)": [76, 80, 75, 70, 66, 67],
    "weight (lbs)": [240, 262, 225, 122, 115, 117],
    "occupation": ["Football Player", "Football Player", "Football Player", "Ballet Dancer", "Ballet Dancer", "Ballet Dancer"]
})
df
```

```python
df['height (cm)'] = round(df['height (in)'] * 2.54, 1)
df['weight (kg)'] = round(df['weight (lbs)'] / 2.205, 1)
df
```

```python
df = df[['occupation', 'height (in)', 'height (cm)', 'weight (lbs)', 'weight (kg)']]
df
```

```python
round(df.groupby('occupation').mean(), 1)
```
